```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math


features = ["Quality","Angle","Distance","CorrectedDistance"]
test = pd.read_csv('realtime_lidar_test.csv', sep='\s+',skiprows=2,names = features) #_v2
#X = test.drop()

#print(test.dtypes)

d = []
for i in range(0,len(test.Quality)):
    #print(test.Quality[i])
    d.append(test.Distance[i]*(math.cos(math.radians(abs(test.Angle[i] - 270)))))

e = pd.Series(d)
#print(test)
#print(e)
    


test["CorrectedDistance"] = e.values

#print(test)


#test["CorrectedDistance"] = d
#print(d)   
#print(test.shape)

#print(test)
#print(test.Quality)

plt.plot(test.CorrectedDistance)
plt.show()
test.to_csv('cleaned.csv', index=False)
```
